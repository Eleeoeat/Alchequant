```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

DB_PATH = ROOT / 'data' / 'stocks.db'
print(f'项目根目录: {ROOT}')
print(f'数据库: {DB_PATH}')
```

```python
from src.factors import calculate_factor_scores, filter_factor_scores, get_candidate_list

factors = calculate_factor_scores(db_path=str(DB_PATH), lookback_days=252, min_obs=120)
print(f'纳入股票: {len(factors)}')
factors[['代码', '名称', '综合评分', '动量评分', '趋势评分', '风险评分', '风险等级', '候选标记']].head(10)
```

```python
ranked = filter_factor_scores(factors, sort_factor='综合评分', risk_filter='全部', top_n=10)
ranked[['代码', '名称', '综合评分', '风险等级', '候选标记']]
```

```python
candidates = get_candidate_list(factors, top_n=10)
candidates[['代码', '名称', '综合评分', '动量60日', '趋势评分', '风险评分', '风险等级']]
```

小结：因子评分只依赖本地行情数据。价格分位是行情代理指标，不等同于财务估值。
